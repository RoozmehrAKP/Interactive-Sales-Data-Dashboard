```python
import pandas as pd

# importing data
file_path = "d:\\Git Project\\Interactive Sales Data Dashboard\\Data\\sales_data_sample.csv"

df = pd.read_csv(file_path, encoding="latin1")

print("Data first 10 rows")
display(df.head(10))

print("Data Describe")
display(df.describe())

print("Data Information")
df.info()
```

### Managing null objects

```python
df["STATE"] = df["STATE"].fillna("Unknown")
df["POSTALCODE"] = df["POSTALCODE"].fillna("00000")
df.drop(columns=["ADDRESSLINE2"], inplace=True)

```

```python
print("Data Information")
df.info()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

num_cols = ["SALES", "QUANTITYORDERED", "PRICEEACH"]
for col in num_cols:
    plt.figure(figsize=(6, 4))
    sns.histplot(df[col], bins=30, kde=True)
    plt.title(f"Distribution of {col}")
    plt.show()
```

```python
import numpy as np

plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Heat Map")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x="PRODUCTLINE", y="SALES", data=df)
plt.xticks(rotation=45)
plt.title("Comparison of sales volume based on product category")
plt.show()
```

```python
df["ORDERDATE"] = pd.to_datetime(df["ORDERDATE"])  # تبدیل تاریخ به datetime
df_monthly_sales = df.groupby("MONTH_ID")["SALES"].sum().reset_index()

plt.figure(figsize=(10, 5))
sns.lineplot(x="MONTH_ID", y="SALES", data=df_monthly_sales, marker="o")
plt.title("The trend of sales changes over the months")
plt.xticks(df_monthly_sales["MONTH_ID"])
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(x="DEALSIZE", y="SALES", data=df)
plt.title(" Sales vs Transaction Size")
plt.show()
```

```python


# تبدیل ستون ORDERDATE به datetime
df["ORDERDATE"] = pd.to_datetime(df["ORDERDATE"])

# مرتب‌سازی داده‌ها بر اساس تاریخ
df = df.sort_values("ORDERDATE")

# ایجاد نمودار روند فروش در طول زمان
plt.figure(figsize=(20, 6))
df.groupby("ORDERDATE")["SALES"].sum().plot(marker="o", linestyle="-", color="b")

# تنظیمات نمودار
plt.title("Sales trend over time")
plt.xlabel("Order Date")
plt.ylabel("Total Sales")
plt.grid(True)
plt.xticks(rotation=45)

# نمایش نمودار
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns



## 📌 ۱. پرفروش‌ترین دسته‌بندی‌ها (PRODUCTLINE)
product_sales = df.groupby("PRODUCTLINE")["SALES"].sum().sort_values(ascending=False)

# رسم نمودار میله‌ای
plt.figure(figsize=(12, 6))
sns.barplot(x=product_sales.index, y=product_sales.values, palette="viridis")
plt.title("Best Selling Product Categories")
plt.xlabel("Product Categories")
plt.ylabel("Total sales")
plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()

## 📌 ۲. پرفروش‌ترین محصولات (PRODUCTCODE)
top_products = df.groupby("PRODUCTCODE")["SALES"].sum().sort_values(ascending=False).head(10)

# رسم نمودار میله‌ای برای ۱۰ محصول پرفروش
plt.figure(figsize=(12, 6))
sns.barplot(x=top_products.index, y=top_products.values, palette="coolwarm")
plt.title("10 Best-Selling Product")
plt.xlabel("Product code")
plt.ylabel("Total sales")
plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

```python
# 📌 ۱. محاسبه مجموع فروش برای هر کشور
country_sales = df.groupby("COUNTRY")["SALES"].sum().sort_values(ascending=False)

# 📊 رسم نمودار میله‌ای برای مقایسه فروش در کشورها
plt.figure(figsize=(14, 6))
sns.barplot(x=country_sales.index, y=country_sales.values, palette="coolwarm")
plt.title("Compare sales in different countries")
plt.xlabel("Country")
plt.ylabel("Total sales")
plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

```python
import plotly.express as px

# حذف مقادیر نال از ستون COUNTRY و SALES
df = df.dropna(subset=["COUNTRY", "SALES"])

# 📌 ۱. محاسبه مجموع فروش برای هر کشور
country_sales = df.groupby("COUNTRY")["SALES"].sum().reset_index()

# 📌 ۲. رسم نقشه Choropleth
fig = px.choropleth(
    country_sales,
    locations="COUNTRY",
    locationmode="country names",
    color="SALES",
    hover_name="COUNTRY",
    color_continuous_scale="Reds",
    title="میزان فروش در کشورهای مختلف"
)

# نمایش نقشه
fig.show()
```

```python
# بررسی مقادیر نال در ستون STATUS و حذف آنها در صورت وجود
df = df.dropna(subset=["STATUS"])

# 📌 ۱. شمارش تعداد سفارشات در هر وضعیت
status_counts = df["STATUS"].value_counts()

# 📊 رسم نمودار ستونی برای توزیع وضعیت سفارشات
plt.figure(figsize=(10, 6))
sns.barplot(x=status_counts.index, y=status_counts.values, palette="coolwarm")
plt.title("Orders status")
plt.xlabel("Status")
plt.ylabel("Orders count")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

```python
# بررسی مقدارهای نال در ستون DEALSIZE و حذف آنها در صورت نیاز
df = df.dropna(subset=["DEALSIZE"])

# 📌 ۱. شمارش تعداد معاملات در هر دسته‌بندی
deal_counts = df["DEALSIZE"].value_counts()

# 📊 رسم نمودار ستونی برای توزیع اندازه معاملات
plt.figure(figsize=(8, 5))
sns.barplot(x=deal_counts.index, y=deal_counts.values, palette="coolwarm")
plt.title("Deal size distribution")
plt.xlabel("Deal size")
plt.ylabel("Deal count")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()

# 📌 ۲. بررسی میانگین فروش بر اساس اندازه معامله
avg_sales_by_dealsize = df.groupby("DEALSIZE")["SALES"].mean().reset_index()

# 📊 رسم نمودار میله‌ای برای مقایسه میانگین فروش در هر اندازه معامله
plt.figure(figsize=(8, 5))
sns.barplot(x="DEALSIZE", y="SALES", data=avg_sales_by_dealsize, palette="viridis")
plt.title("Average sales per Deal size")
plt.xlabel("Deal size")
plt.ylabel("Average Sales")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```
