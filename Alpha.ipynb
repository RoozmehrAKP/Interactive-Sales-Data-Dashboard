```python
import pandas as pd

# importing data
file_path = "d:\\Git Project\\Interactive Sales Data Dashboard\\Data\\sales_data_sample.csv"

df = pd.read_csv(file_path, encoding="latin1")

print("Data first 10 rows")
display(df.head(10))

print("Data Describe")
display(df.describe())

print("Data Information")
df.info()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns



## ðŸ“Œ Û±. Ù¾Ø±ÙØ±ÙˆØ´â€ŒØªØ±ÛŒÙ† Ø¯Ø³ØªÙ‡â€ŒØ¨Ù†Ø¯ÛŒâ€ŒÙ‡Ø§ (PRODUCTLINE)
product_sales = df.groupby("PRODUCTLINE")["SALES"].sum().sort_values(ascending=False)

# Ø±Ø³Ù… Ù†Ù…ÙˆØ¯Ø§Ø± Ù…ÛŒÙ„Ù‡â€ŒØ§ÛŒ
plt.figure(figsize=(12, 6))
sns.barplot(x=product_sales.index, y=product_sales.values, palette="viridis")
plt.title("Best Selling Product Categories")
plt.xlabel("Product Categories")
plt.ylabel("Total sales")
plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()

## ðŸ“Œ Û². Ù¾Ø±ÙØ±ÙˆØ´â€ŒØªØ±ÛŒÙ† Ù…Ø­ØµÙˆÙ„Ø§Øª (PRODUCTCODE)
top_products = df.groupby("PRODUCTCODE")["SALES"].sum().sort_values(ascending=False).head(10)

# Ø±Ø³Ù… Ù†Ù…ÙˆØ¯Ø§Ø± Ù…ÛŒÙ„Ù‡â€ŒØ§ÛŒ Ø¨Ø±Ø§ÛŒ Û±Û° Ù…Ø­ØµÙˆÙ„ Ù¾Ø±ÙØ±ÙˆØ´
plt.figure(figsize=(12, 6))
sns.barplot(x=top_products.index, y=top_products.values, palette="coolwarm")
plt.title("10 Best-Selling Product")
plt.xlabel("Product code")
plt.ylabel("Total sales")
plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

```python
# ðŸ“Œ Û±. Ù…Ø­Ø§Ø³Ø¨Ù‡ Ù…Ø¬Ù…ÙˆØ¹ ÙØ±ÙˆØ´ Ø¨Ø±Ø§ÛŒ Ù‡Ø± Ú©Ø´ÙˆØ±
country_sales = df.groupby("COUNTRY")["SALES"].sum().sort_values(ascending=False)

# ðŸ“Š Ø±Ø³Ù… Ù†Ù…ÙˆØ¯Ø§Ø± Ù…ÛŒÙ„Ù‡â€ŒØ§ÛŒ Ø¨Ø±Ø§ÛŒ Ù…Ù‚Ø§ÛŒØ³Ù‡ ÙØ±ÙˆØ´ Ø¯Ø± Ú©Ø´ÙˆØ±Ù‡Ø§
plt.figure(figsize=(14, 6))
sns.barplot(x=country_sales.index, y=country_sales.values, palette="coolwarm")
plt.title("Compare sales in different countries")
plt.xlabel("Country")
plt.ylabel("Total sales")
plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

```python
import plotly.express as px

# Ø­Ø°Ù Ù…Ù‚Ø§Ø¯ÛŒØ± Ù†Ø§Ù„ Ø§Ø² Ø³ØªÙˆÙ† COUNTRY Ùˆ SALES
df = df.dropna(subset=["COUNTRY", "SALES"])

# ðŸ“Œ Û±. Ù…Ø­Ø§Ø³Ø¨Ù‡ Ù…Ø¬Ù…ÙˆØ¹ ÙØ±ÙˆØ´ Ø¨Ø±Ø§ÛŒ Ù‡Ø± Ú©Ø´ÙˆØ±
country_sales = df.groupby("COUNTRY")["SALES"].sum().reset_index()

# ðŸ“Œ Û². Ø±Ø³Ù… Ù†Ù‚Ø´Ù‡ Choropleth
fig = px.choropleth(
    country_sales,
    locations="COUNTRY",
    locationmode="country names",
    color="SALES",
    hover_name="COUNTRY",
    color_continuous_scale="Reds",
    title="Ù…ÛŒØ²Ø§Ù† ÙØ±ÙˆØ´ Ø¯Ø± Ú©Ø´ÙˆØ±Ù‡Ø§ÛŒ Ù…Ø®ØªÙ„Ù"
)

# Ù†Ù…Ø§ÛŒØ´ Ù†Ù‚Ø´Ù‡
fig.show()
```

```python
# Ø¨Ø±Ø±Ø³ÛŒ Ù…Ù‚Ø§Ø¯ÛŒØ± Ù†Ø§Ù„ Ø¯Ø± Ø³ØªÙˆÙ† STATUS Ùˆ Ø­Ø°Ù Ø¢Ù†Ù‡Ø§ Ø¯Ø± ØµÙˆØ±Øª ÙˆØ¬ÙˆØ¯
df = df.dropna(subset=["STATUS"])

# ðŸ“Œ Û±. Ø´Ù…Ø§Ø±Ø´ ØªØ¹Ø¯Ø§Ø¯ Ø³ÙØ§Ø±Ø´Ø§Øª Ø¯Ø± Ù‡Ø± ÙˆØ¶Ø¹ÛŒØª
status_counts = df["STATUS"].value_counts()

# ðŸ“Š Ø±Ø³Ù… Ù†Ù…ÙˆØ¯Ø§Ø± Ø³ØªÙˆÙ†ÛŒ Ø¨Ø±Ø§ÛŒ ØªÙˆØ²ÛŒØ¹ ÙˆØ¶Ø¹ÛŒØª Ø³ÙØ§Ø±Ø´Ø§Øª
plt.figure(figsize=(10, 6))
sns.barplot(x=status_counts.index, y=status_counts.values, palette="coolwarm")
plt.title("Orders status")
plt.xlabel("Status")
plt.ylabel("Orders count")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

```python
# Ø¨Ø±Ø±Ø³ÛŒ Ù…Ù‚Ø¯Ø§Ø±Ù‡Ø§ÛŒ Ù†Ø§Ù„ Ø¯Ø± Ø³ØªÙˆÙ† DEALSIZE Ùˆ Ø­Ø°Ù Ø¢Ù†Ù‡Ø§ Ø¯Ø± ØµÙˆØ±Øª Ù†ÛŒØ§Ø²
df = df.dropna(subset=["DEALSIZE"])

# ðŸ“Œ Û±. Ø´Ù…Ø§Ø±Ø´ ØªØ¹Ø¯Ø§Ø¯ Ù…Ø¹Ø§Ù…Ù„Ø§Øª Ø¯Ø± Ù‡Ø± Ø¯Ø³ØªÙ‡â€ŒØ¨Ù†Ø¯ÛŒ
deal_counts = df["DEALSIZE"].value_counts()

# ðŸ“Š Ø±Ø³Ù… Ù†Ù…ÙˆØ¯Ø§Ø± Ø³ØªÙˆÙ†ÛŒ Ø¨Ø±Ø§ÛŒ ØªÙˆØ²ÛŒØ¹ Ø§Ù†Ø¯Ø§Ø²Ù‡ Ù…Ø¹Ø§Ù…Ù„Ø§Øª
plt.figure(figsize=(8, 5))
sns.barplot(x=deal_counts.index, y=deal_counts.values, palette="coolwarm")
plt.title("Deal size distribution")
plt.xlabel("Deal size")
plt.ylabel("Deal count")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()

# ðŸ“Œ Û². Ø¨Ø±Ø±Ø³ÛŒ Ù…ÛŒØ§Ù†Ú¯ÛŒÙ† ÙØ±ÙˆØ´ Ø¨Ø± Ø§Ø³Ø§Ø³ Ø§Ù†Ø¯Ø§Ø²Ù‡ Ù…Ø¹Ø§Ù…Ù„Ù‡
avg_sales_by_dealsize = df.groupby("DEALSIZE")["SALES"].mean().reset_index()

# ðŸ“Š Ø±Ø³Ù… Ù†Ù…ÙˆØ¯Ø§Ø± Ù…ÛŒÙ„Ù‡â€ŒØ§ÛŒ Ø¨Ø±Ø§ÛŒ Ù…Ù‚Ø§ÛŒØ³Ù‡ Ù…ÛŒØ§Ù†Ú¯ÛŒÙ† ÙØ±ÙˆØ´ Ø¯Ø± Ù‡Ø± Ø§Ù†Ø¯Ø§Ø²Ù‡ Ù…Ø¹Ø§Ù…Ù„Ù‡
plt.figure(figsize=(8, 5))
sns.barplot(x="DEALSIZE", y="SALES", data=avg_sales_by_dealsize, palette="viridis")
plt.title("Average sales per Deal size")
plt.xlabel("Deal size")
plt.ylabel("Average Sales")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```
