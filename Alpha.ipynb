- **Whiskers:** Indicate the minimum and maximum non-outlier values.  
- **Outliers:** Points beyond the whiskers are considered **outliers**, representing unusual or extreme values.

# **Plotting Outlier Orders Over Time**  

To determine whether these outlier orders are concentrated within specific time periods (e.g., holidays, end of the fiscal year, special discounts), we can:  

1. Convert the `ORDERDATE` column to `datetime` format.  
2. Filter orders identified as outliers.  
3. Aggregate outlier sales by month or week.  
4. Plot a **time series chart (lineplot or barplot)** to visualize trends.  

This will help identify whether these unusual orders align with seasonal patterns or special events.  


````python
# Adding a month column to outlier data
outliers['ORDER_MONTH'] = outliers['ORDERDATE'].dt.to_period('M')

# Counting the number of outlier orders per month
outlier_counts = outliers.groupby('ORDER_MONTH').size()

plt.figure(figsize=(12, 5))
outlier_counts.plot(kind='bar', color='red')
plt.title('Number of Outlier Sales Per Month')
plt.xlabel('Month')
plt.ylabel('Number of Outlier Orders')
plt.xticks(rotation=45)
plt.show()
````

````python

# Calculate Q1, Q3, and IQR for SALES
Q1 = df['SALES'].quantile(0.25)
Q3 = df['SALES'].quantile(0.75)
IQR = Q3 - Q1

# Define the upper bound (keep lower bound untouched)
upper_bound = Q3 + 1.5 * IQR

# Filter the dataset
df_cleaned = df[df['SALES'] <= upper_bound]  # Keeping lower outliers
df_outliers = df[df['SALES'] > upper_bound]  # Only upper outliers

# Save cleaned data and outliers separately
df_cleaned.to_csv("cleaned_Outlaiers_data.csv", index=False)
df_outliers.to_csv("sales_outliers.csv", index=False)

print(f"Original dataset size: {len(df)}")
print(f"Cleaned dataset size (without upper outliers): {len(df_cleaned)}")
print(f"Outliers dataset size: {len(df_outliers)}")
````
