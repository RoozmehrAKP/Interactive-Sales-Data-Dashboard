```python
import pandas as pd

# importing data
file_path = "d:\\Git Project\\Interactive Sales Data Dashboard\\Data\\sales_data_sample.csv"

df = pd.read_csv(file_path, encoding="latin1")

print("Data first 10 rows")
display(df.head(10))

print("Data Describe")
display(df.describe())

print("Data Information")
df.info()
```

### Managing null objects

```python
df["STATE"] = df["STATE"].fillna("Unknown")
df["POSTALCODE"] = df["POSTALCODE"].fillna("00000")
df.drop(columns=["ADDRESSLINE2"], inplace=True)

```

```python
print("Data Information")
df.info()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

num_cols = ["SALES", "QUANTITYORDERED", "PRICEEACH"]
for col in num_cols:
    plt.figure(figsize=(6, 4))
    sns.histplot(df[col], bins=30, kde=True)
    plt.title(f"Distribution of {col}")
    plt.show()
```

```python
import numpy as np

plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Heat Map")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x="PRODUCTLINE", y="SALES", data=df)
plt.xticks(rotation=45)
plt.title("Comparison of sales volume based on product category")
plt.show()
```

```python
df["ORDERDATE"] = pd.to_datetime(df["ORDERDATE"])  # ØªØ¨Ø¯ÛŒÙ„ ØªØ§Ø±ÛŒØ® Ø¨Ù‡ datetime
df_monthly_sales = df.groupby("MONTH_ID")["SALES"].sum().reset_index()

plt.figure(figsize=(10, 5))
sns.lineplot(x="MONTH_ID", y="SALES", data=df_monthly_sales, marker="o")
plt.title("The trend of sales changes over the months")
plt.xticks(df_monthly_sales["MONTH_ID"])
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(x="DEALSIZE", y="SALES", data=df)
plt.title(" Sales vs Transaction Size")
plt.show()
```
