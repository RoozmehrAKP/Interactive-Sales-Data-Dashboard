```python
import pandas as pd

# importing data
file_path = "d:\\Git Project\\Interactive Sales Data Dashboard\\Data\\Cleaned_data.csv"

df = pd.read_csv(file_path, encoding="latin1")

print("Data first 10 rows")
display(df.head(10))

print("Data Describe")
display(df.describe())

print("Data Information")
df.info()
```

```python
mean_sales = df['SALES'].mean()
std_sales = df['SALES'].std()

# شناسایی فروش‌های پرت
outlier_threshold = mean_sales + 2.5 * std_sales
outliers = df[df['SALES'] > outlier_threshold]

print(outliers)
```

```python
# تبدیل ORDERDATE به نوع تاریخ
df['ORDERDATE'] = pd.to_datetime(df['ORDERDATE'])
```

```python
# استخراج ماه و افزودن آن به DataFrame
df['Order Month'] = df['ORDERDATE'].dt.month_name()  # برای نام ماه
# یا
# df['Order Month'] = df['ORDERDATE'].dt.month  # برای شماره ماه
```

```python
# تجزیه و تحلیل فروش بر اساس ماه
monthly_sales = df.groupby('Order Month')['SALES'].sum().reset_index()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# ایجاد نمودار
plt.figure(figsize=(12, 6))
sns.barplot(x='Order Month', y='SALES', data=monthly_sales,)
plt.title('Total Sales per Month', fontsize=20, color='gray', loc='left', fontweight='bold')
plt.xlabel('Order Month', fontsize=12, loc='left', color='gray')
plt.ylabel('SALES', fontsize=12, loc='bottom', color='gray')

plt.xticks(rotation=45)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['top'].set_visible(False)
plt.tick_params(colors='gray')


plt.show()
```

```python

norm = plt.Normalize(monthly_sales["SALES"].min(), monthly_sales["SALES"].max())
colors = plt.cm.cividis(norm(monthly_sales["SALES"]))  # تغییر رنگ با cmap مختلف

# رسم نمودار
plt.figure(figsize=(12, 6))
sns.barplot(x='Order Month', y='SALES', data=monthly_sales, palette=colors)

# تنظیمات نمودار
plt.title('Total Sales per Month', fontsize=20, color='gray', loc='left', fontweight='bold')
plt.xlabel('Order Month', fontsize=12, loc='left', color='gray')
plt.ylabel('SALES', fontsize=12, loc='bottom', color='gray')

plt.xticks(rotation=45)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['top'].set_visible(False)
plt.tick_params(colors='gray')


plt.show()
```

```python
# رسم نمودار
fig, ax1 = plt.subplots(figsize=(12, 6))

# رسم نمودار میله‌ای
sns.barplot(x='Order Month', y='SALES', data=monthly_sales, palette=colors, ax=ax1)
ax1.set_ylabel("Total Sales", fontsize=10, color='gray')
ax1.set_xlabel("Order Month", fontsize=10, color='gray')
ax1.set_title("Total Sales per Month with Trend Line", fontsize=16, color='gray', fontweight='bold', loc='left')

# ایجاد محور دوم برای نمودار خطی
ax2 = ax1.twinx()
sns.lineplot(x='Order Month', y='SALES', data=monthly_sales, color='red', marker='o', linewidth=2, ax=ax2)
ax2.set_ylabel("Sales Trend", fontsize=12, color='red')

# تنظیمات ظاهری
plt.xticks(rotation=45)
ax1.spines['right'].set_visible(False)
ax1.spines['top'].set_visible(False)
ax2.spines['right'].set_color('red')  # تاکید روی محور نمودار خطی
ax1.tick_params(colors='gray')
ax2.tick_params(colors='red')
plt.gca().spines['top'].set_visible(False)
plt.show()
```

```python
top_customers = df.groupby('CUSTOMERNAME')['SALES'].sum().nlargest(10).reset_index()
colors2 = plt.cm.cividis_r(norm(monthly_sales["SALES"]))

plt.figure(figsize=(12, 6))
sns.barplot(x='SALES', y='CUSTOMERNAME', data=top_customers, palette=colors2)
plt.title('Top 10 Customers by Sales')
plt.show()
```
