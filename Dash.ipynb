```python
import pandas as pd

# importing data
file_path = "d:\\Git Project\\Interactive Sales Data Dashboard\\Data\\Cleaned_data.csv"

df = pd.read_csv(file_path, encoding="latin1")

print("Data first 10 rows")
display(df.head(10))

print("Data Describe")
display(df.describe())

print("Data Information")
df.info()
```

```python
mean_sales = df['SALES'].mean()
std_sales = df['SALES'].std()

# شناسایی فروش‌های پرت
outlier_threshold = mean_sales + 2.5 * std_sales
outliers = df[df['SALES'] > outlier_threshold]

print(outliers)
```

```python
# تبدیل ORDERDATE به نوع تاریخ
df['ORDERDATE'] = pd.to_datetime(df['ORDERDATE'])
```

```python
# تبدیل ORDERDATE به نوع تاریخ
outliers['ORDERDATE'] = pd.to_datetime(outliers['ORDERDATE'])
```

```python
# استخراج ماه و افزودن آن به DataFrame
df['Order Month'] = df['ORDERDATE'].dt.month_name()  # برای نام ماه
# یا
# df['Order Month'] = df['ORDERDATE'].dt.month  # برای شماره ماه
```

```python
# تجزیه و تحلیل فروش بر اساس ماه
monthly_sales = df.groupby('Order Month')['SALES'].sum().reset_index()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python

norm = plt.Normalize(monthly_sales["SALES"].min(), monthly_sales["SALES"].max())
colors = list(plt.cm.cividis(norm(monthly_sales["SALES"])))  # تبدیل آرایه به لیست

# رسم نمودار
fig, ax1 = plt.subplots(figsize=(12, 6))

# رسم نمودار میله‌ای
sns.barplot(x='Order Month', y='SALES', data=monthly_sales, palette="cividis_r", ax=ax1,hue='Order Month')
ax1.set_ylabel("Total Sales", fontsize=10, color='gray')
ax1.set_xlabel("Order Month", fontsize=10, color='gray')
ax1.set_title("Total Sales per Month with Trend Line", fontsize=16, color='gray', fontweight='bold', loc='left')

# ایجاد محور دوم برای نمودار خطی
ax2 = ax1.twinx()
sns.lineplot(x='Order Month', y='SALES', data=monthly_sales, color='red', marker='o', linewidth=2, ax=ax2)
ax2.set_ylabel("Sales Trend", fontsize=12, color='red')

# تنظیمات ظاهری
plt.xticks(rotation=45)
ax1.spines['right'].set_visible(False)
ax1.spines['top'].set_visible(False)
ax2.spines['right'].set_color('red')  # تاکید روی محور نمودار خطی
ax1.tick_params(colors='gray')
ax2.tick_params(colors='red')
plt.gca().spines['top'].set_visible(False)
plt.show()
```

```python
# محاسبه مجموع فروش برای هر مشتری
customer_sales = df.groupby('CUSTOMERNAME')['SALES'].sum().reset_index()

# ایجاد هیستوگرام توزیع فروش
plt.figure(figsize=(12, 6))
sns.histplot(customer_sales['SALES'], bins=30, kde=True)
plt.title('Distribution of Sales among Customers', fontsize=16, color='gray', fontweight='bold', loc='left')
plt.xlabel('Sales Amount', fontsize=10, color='gray')
plt.ylabel('Frequency', fontsize=10, color='gray')
plt.axvline(x=customer_sales['SALES'].mean(), color='r', linestyle='--', label='Mean Sales')
plt.axvline(x=customer_sales['SALES'].mean() + 2.5 * customer_sales['SALES'].std(), color='y', linestyle='--', label='Outlier Threshold')
plt.legend()
plt.xticks(color='gray')
plt.yticks(color='gray')
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

plt.show()
```
