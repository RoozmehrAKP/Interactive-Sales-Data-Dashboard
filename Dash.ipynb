# Identifying Outlier Sales
Calculates the mean and standard deviation of the 'SALES' column, then identifies and prints outlier sales.

# Identify outlier sales (those above a threshold of 2.5 standard deviations from the mean)


````python
mean_sales = df['SALES'].mean()
std_sales = df['SALES'].std()

# Detecting Outliers
outlier_threshold = mean_sales + 2.5 * std_sales
outliers = df[df['SALES'] > outlier_threshold]

print(outliers)
````

````python
# Define ordered list of months
ordered_months = ['January', 'February', 'March', 'April', 'May', 'June', 
                  'July', 'August', 'September', 'October', 'November', 'December']

# Convert 'Order Month' to a categorical type with ordered months
monthly_sales['Order Month'] = pd.Categorical(monthly_sales['Order Month'], categories=ordered_months, ordered=True)

# Normalize sales values for coloring
norm = plt.Normalize(monthly_sales["SALES"].min(), monthly_sales["SALES"].max())
colors = list(plt.cm.cividis(norm(monthly_sales["SALES"])))  # changing array into list

fig, ax1 = plt.subplots(figsize=(12, 6))

# Create the bar plot
sns.barplot(x='Order Month', y='SALES', data=monthly_sales, palette="cividis_r", ax=ax1, hue='Order Month')
ax1.set_ylabel("Total Sales", fontsize=10, color='gray')
ax1.set_xlabel("Order Month", fontsize=10, color='gray')
ax1.set_title("Total Sales per Month with Trend Line", fontsize=16, color='gray', fontweight='bold', loc='left')

# Create a secondary axis for the line plot
ax2 = ax1.twinx()
sns.lineplot(x='Order Month', y='SALES', data=monthly_sales, color='red', marker='o', linewidth=2, ax=ax2)
ax2.set_ylabel("Sales Trend", fontsize=12, color='red')

plt.xticks(rotation=45)
ax1.spines['right'].set_visible(False)
ax1.spines['top'].set_visible(False)
ax2.spines['right'].set_color('red')  
ax1.tick_params(colors='gray')
ax2.tick_params(colors='red')
plt.gca().spines['top'].set_visible(False)
plt.show()
````

````python
# تعیین آستانه Outlier (بر اساس چارک 95٪)
threshold = df['SALES'].quantile(0.75)

# اضافه کردن ستون جدید برای مشخص کردن اینکه هر رکورد Outlier هست یا نه
df['Outlier'] = (df['SALES'] > threshold).astype(int)

# استخراج سال از تاریخ سفارش
df['Year'] = pd.to_datetime(df['ORDERDATE']).dt.year

# محاسبه مجموع فروش و فروش Outlier در هر سال
yearly_sales = df.groupby('Year')['SALES'].sum().reset_index()
yearly_outlier_sales = df[df['Outlier'] == 1].groupby('Year')['SALES'].sum().reset_index()

# ادغام داده‌ها برای مقایسه فروش کلی و فروش Outlier
yearly_sales = yearly_sales.merge(yearly_outlier_sales, on="Year", how="left", suffixes=("_Total", "_Outlier"))
yearly_sales.fillna(0, inplace=True)  # جایگذاری مقادیر NaN با صفر

# رسم نمودار
plt.figure(figsize=(12, 6))
sns.barplot(x='Year', y='SALES_Total', data=yearly_sales, color='gray', label="Total Sales")
sns.barplot(x='Year', y='SALES_Outlier', data=yearly_sales, color='yellow', label="Outlier Sales", alpha=0.5)  

# تنظیمات ظاهری نمودار
plt.title('Annual Sales with Outlier Contribution', fontsize=16, color='gray', fontweight='bold', loc='left')
plt.xlabel('Year', fontsize=10, color='gray')
plt.ylabel('Sales Amount', fontsize=10, color='gray')
plt.legend()

plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(color='gray')
plt.yticks(color='gray')
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

plt.show()
````
