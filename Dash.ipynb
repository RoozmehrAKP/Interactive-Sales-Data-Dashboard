# Identifying Outlier Sales
Calculates the mean and standard deviation of the 'SALES' column, then identifies and prints outlier sales.

# Identify outlier sales (those above a threshold of 2.5 standard deviations from the mean)


````python
mean_sales = df['SALES'].mean()
std_sales = df['SALES'].std()

# Detecting Outliers
outlier_threshold = mean_sales + 2.5 * std_sales
outliers = df[df['SALES'] > outlier_threshold]

print(outliers)
````

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming df is already loaded and contains ORDERDATE and SALES

# Create Year and Month fields
df['Year'] = df['ORDERDATE'].dt.year
df['Month'] = df['ORDERDATE'].dt.month_name(locale='en_US.UTF-8')  # Get month names in English

# Define ordered list of months
ordered_months = ['January', 'February', 'March', 'April', 'May', 'June', 
                  'July', 'August', 'September', 'October', 'November', 'December']

# Convert Month to a categorical type with ordered months
df['Month'] = pd.Categorical(df['Month'], categories=ordered_months, ordered=True)

# Calculate total sales by Year and Month
monthly_sales = df.groupby(['Year', 'Month'])['SALES'].sum().reset_index()

# Create a vertical bar plot to show seasonal patterns
plt.figure(figsize=(12, 6))
sns.barplot(data=monthly_sales, x='Month', y='SALES', hue='Year', palette='cividis')  # Change bar colors to cividis
plt.title('Sales Distribution by Month', fontsize=16, color='gray', fontweight='bold', loc='left')
plt.xticks(rotation=45)
plt.xticks(color='gray')
plt.yticks(color='gray')
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.xlabel('Month', fontsize=10, color='gray')
plt.ylabel('SALES', fontsize=10, color='gray')
plt.show()

# Create a vertical line plot
plt.figure(figsize=(12, 6))
sns.lineplot(data=monthly_sales, x='Month', y='SALES', hue='Year', palette='cividis')  # Change line colors to cividis
plt.title('Monthly Sales Trend', fontsize=16, color='gray', fontweight='bold', loc='left')
plt.xticks(rotation=45)
plt.xticks(color='gray')
plt.yticks(color='gray')
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.xlabel('Month', fontsize=10, color='gray')
plt.ylabel('SALES', fontsize=10, color='gray')
plt.show()
````

# Comparing Different Product Categories

````python

# Group sales by product category
product_sales = df.groupby('PRODUCTLINE')['SALES'].sum().reset_index()

# Sort the categories by sales in descending order
product_sales = product_sales.sort_values(by='SALES', ascending=False)

# Identify best and worst performing categories
best_performance = product_sales.loc[product_sales['SALES'].idxmax()]
worst_performance = product_sales.loc[product_sales['SALES'].idxmin()]

print("Best Performing PRODUCTLINE:")
print(best_performance)
print("Worst Performing PRODUCTLINE:")
print(worst_performance)

# Visualize sales by product category
plt.figure(figsize=(12, 6))
sns.barplot(x='PRODUCTLINE', y='SALES', data=product_sales, palette='cividis')
plt.title('Sales by PRODUCTLINE', fontsize=16, color='gray', fontweight='bold', loc='left')
plt.xticks(rotation=45)
plt.ylabel('Total Sales')
plt.xlabel('Product Category')
plt.xticks(color='gray')
plt.yticks(color='gray')
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.xlabel('PRODUCTLINE', fontsize=10, color='gray')
plt.ylabel('SALES', fontsize=10, color='gray')
plt.show()
````
